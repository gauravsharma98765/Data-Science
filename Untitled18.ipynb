```python
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
driver = webdriver.Chrome('C:/Users/lenovo/Downloads/chromedriver')
```

```python
Brand = []
Rating = []
Description = []
Offer_Price = []
MRP = []
Discount_Percentage = []
```

```python
for i in range(1,31):
    driver.get('https://www.myntra.com/men-shoes?p={}'.format(i))
    html = driver.page_source
    soup = BeautifulSoup(html,"html.parser")
    shoe_div = soup.find_all('li', class_='product-base')
    time.sleep(2)
    
    for container in shoe_div :
        rate = container.find('div', class_='product-ratingsContainer').span.text if container.find('div', class_='product-ratingsContainer') else "-"
        Rating.append(rate)

        brand = container.find('h3', class_='product-brand').text if container.find('h3', class_='product-brand') else "-"
        Brand.append(brand)

        desc = container.find('h4', class_='product-product').text if container.find('h4', class_='product-product') else "-"
        Description.append(desc)

        price = container.find('span', class_='product-discountedPrice').text if container.find('span', class_='product-discountedPrice') else "-"
        Offer_Price.append(price)

        prc = container.find('span', class_='product-strike').text if container.find('span', class_='product-strike') else "-"
        MRP.append(prc)

        disc_perc = container.find('span', class_='product-discountPercentage').text if container.find('span', class_='product-discountPercentage') else "-"
        Discount_Percentage.append(disc_perc)
```

```python
Shoe = pd.DataFrame({
    'Brand' : Brand,
    'Description' : Description,
    'Offer Price' : Offer_Price,
    'MRP' : MRP,
    'Discount Percent' :Discount_Percentage,
    'Rating' : Rating
})
```

```python
Shoe.to_csv('Shoe.csv')
```
